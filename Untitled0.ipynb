```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error # Corrected 'mean_squared error'
data=pd.read_excel("/content/Student _csv.xlsx")
X=data[['Hours']]
y=data['Scores '] # Corrected column name: 'Scores ' instead of 'Scores'
model=LinearRegression()
model.fit(X,y)
predicted_score=model.predict(X)
#evaluate
mae=mean_absolute_error(y,predicted_score)
mse=mean_squared_error(y,predicted_score)
rmse=np.sqrt(mse)
#show results
print ("Means absolute error:", mae)
print ("Mean squared error:", mse)
print ("Root mean squared error:", rmse)
```
